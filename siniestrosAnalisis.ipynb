* Novedad : Número identificador del siniestro de tránsito.

### **2. Realiza los pasos mediante las técnicas presentadas para evaluar la calidad de los datos. ¿Consideras que tienes datos de calidad?**

```python
# 2. Realiza los pasos mediante las técnicas presentadas para evaluar la calidad de los datos. ¿Consideras que tienes datos de calidad?

#Renombramos las variables con el formato recomendado
def clean_column_name(col):
    # Cambiamos los nombres de las variables reemplazando espacios por _
    col = col.strip().replace(" ", "_")
    return col

# Aplicamos la función a los nombres de las columnas
dataUy.columns = [clean_column_name(col) for col in dataUy.columns]
print(dataUy.columns)
```

```python
# Remplazamos las variables que contenian el caracter "�" por el correcto
dataUy.rename(columns={'Usa_cinturI�n': 'Usa_cinturon','DI�a_de_la_semana': 'Dia_de_la_semana'}, inplace=True)
print(dataUy.columns)

```

```python
# Eliminamos los espacios de toda la base de datos
dataUy = dataUy.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# Remplazamos el caracter "�" en todas las celdas de la base de datos
dataUy = dataUy.applymap(lambda x: x.replace("�", "") if isinstance(x, str) else x)
dataUy.head(10)
```

```python
#VARIABLE NUMÉRICA
#Analizamos que los datos faltantes no aparecen como NaN ni como None, sino que dicen "SIN DATOS".
# Remplazamos el valor "SIN DATOS" de la variable "EDAD" por NaN, y lo convertimos a FLOAT
dataUy['Edad'] = dataUy['Edad'].replace('SIN DATOS',  np.nan)
dataUy['Edad'] = dataUy['Edad'].astype(float)
dataUy['Edad'].head()
```

```python
#VARIABLE DATE
#Convertimos las variables "FECHA" y "HORA" a date
dataUy['Fecha'] = pd.to_datetime(dataUy['Fecha'], format='mixed')
dataUy['Fecha'].tail()

dataUy['Hora'] = pd.to_datetime(dataUy['Hora'], format='%H').dt.time
dataUy['Hora'].head()

#Creamos una nueva variable apartir de la fecha y hora
dataUy['Fecha_Hora'] = dataUy.apply(lambda row: pd.Timestamp.combine(row['Fecha'], row['Hora']), axis=1)

print(dataUy[['Fecha', 'Hora', 'Fecha_Hora']].head())
```

```python
# Nos aseguramos que la columna Hora esté en datetime
dataUy['Hora'] = pd.to_numeric(dataUy['Hora'], errors='coerce')

# Sobreescribimos la variable HORA
dataUy['Hora'] = dataUy['Fecha_Hora'].dt.hour

# Clasificamos los momento del día
def clasificar_momento(hora):
    if 0 <= hora < 6:
        return 'Madrugada'
    elif 6 <= hora < 12:
        return 'Mañana'
    elif 12 <= hora < 18:
        return 'Tarde'
    else:
        return 'Noche'

dataUy['Momento_dia'] = dataUy['Hora'].apply(clasificar_momento)
dataUy['Momento_dia'].describe()
```

```python
print(dataUy['Momento_dia'].unique())
```

```python
# Creamos la variable binaria= 1 si es hora pico, 0 si no lo es.
dataUy['Hora_pico'] = dataUy['Hora'].apply(lambda h: 1 if (6 <= h < 9) or (17 <= h < 19) else 0)
dataUy['Hora_pico'].describe()
```

```python
# Unimos las columnas de latitud (Y) y longitud (X) en una nueva columna llamada 'Coordenadas'
# Para mas tarde en el analisis podemos graficar los puntos en un mapa usando librerías como folium o plotly

dataUy["Coordenadas"] = list(zip(dataUy["Y"], dataUy["X"]))
dataUy.head()
```

```python
#VARIABLES CATEGORICAS
#Remplazamos el valor "SIN DATOS" de las variables "Usa_casco", "Calle", "Localidad", "Tipo_de_Vehiculo", "Zona", "Usa_cinturon" por NONE
columns_to_replace = ['Usa_casco', 'Calle', 'Localidad', 'Tipo_de_Vehiculo', 'Zona','Usa_cinturon', 'Sexo']

for col in columns_to_replace:
    dataUy[col] = dataUy[col].replace('SIN DATOS', None)

# Verficamos que los cambios aplicados anteriormente se ejecutaron correctamente
dataUy.info()
```

```python
#Convertimos la variable "NOVEDAD" en string, es un identificador y no tiene sentido que sea float
dataUy['Novedad'] = dataUy['Novedad'].astype(str)

dataUy.info()
```

```python
# Función para la identificación y conteo de datos nulos
def data_nulls(df):
  null_conteo = df.isnull().sum()
  null_porcen = (df.isnull().sum() / len(df)) * 100

  # Combine into a single DataFrame
  null_report_df = pd.DataFrame({'Null Count': null_conteo,
                              'Null Percentage': null_porcen})
  return(null_report_df)
```

```python
def limpiar_columna_capitalizada(nombre):
    nombre = nombre.strip().lower()

    # Reemplazar tildes y caracteres especiales comunes
    reemplazos = {
        'á': 'a',
        'é': 'e',
        'í': 'i',
        'ó': 'o',
        'ú': 'u',
        'ñ': 'n',
        'ü': 'u'
    }
    for original, reemplazo in reemplazos.items():
        nombre = nombre.replace(original, reemplazo)

    # Eliminar otros caracteres especiales
    for char in ['"', "'", "(", ")", "/", "\\", ".", ",", "¡", "!", "¿", "�", "%", "#", "@", "&", "*", ":", ";", "-"]:
        nombre = nombre.replace(char, '')

    # Reemplazar espacios con guión bajo
    nombre = nombre.replace(' ', '_')

    # Capitalizar solo la primera letra
    if nombre:
        nombre = nombre[0].upper() + nombre[1:]

    return nombre

# Aplicar a todas las columnas del DataFrame
df.columns = [limpiar_columna_capitalizada(col) for col in df.columns]

# Verificar resultado
print(df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Elegimos un estilo y paleta compatible con el daltonismo
sns.set(style="whitegrid", palette="colorblind")

color = sns.color_palette("colorblind")[0]  # azul fuerte accesible

# Creamos el histograma de la variable "EDAD"
plt.figure(figsize=(8, 4))
sns.histplot(df['Edad'], bins=30, kde=True, color=color)

# Le ponemos etiquetas y título
plt.title('Distribución de Edad en Siniestros', fontsize=16)
plt.xlabel('Edad')
plt.ylabel('Frecuencia')

plt.show()
```

El histograma muestra la distribución de la edad de las personas involucradas en siniestros de tránsito. Observamos un pico pronunciado alrededor de los 20 años, lo cual podría estar relacionado con que en Uruguay la edad legal para conducir comienza a los 18 años. Esto implica que, a partir de esa edad, las personas comienzan a participar más activamente en el tránsito como conductores o acompañantes.

A medida que la edad avanza, la frecuencia de participación en siniestros disminuye. Esto puede deberse a diversos factores, como menor exposición al riesgo (menos horas en la vía pública), cambios en hábitos de movilidad, o incluso una conducción más cautelosa.

La distribución general tiene forma de campana con un sesgo hacia la derecha (asimetría positiva), lo cual es típico cuando hay una concentración de casos en edades jóvenes y una disminución progresiva hacia edades más avanzadas.

* La moda se encuentra entre los 20 y 30 años.

#### 2.3 Tablas de Contingencia y Mapas de Calor
Para analizar la relación entre variables categóricas, utilizamos tablas de contingencia visualizadas como mapas de calor. En estos gráficos, los colores indican la frecuencia relativa: los tonos rojos representan combinaciones con más observaciones, mientras que los tonos verdes indican menor frecuencia. Además, se incluyen los porcentajes en cada celda para facilitar la interpretación cuantitativa.

```python
# Primer gráfico: 'ROL' vs 'ZONA'
fig, ax = plt.subplots(figsize=(8, 6))
contingency = pd.crosstab(df['Rol'], df['Zona'])
percentage = contingency / contingency.values.sum() * 100
sns.heatmap(percentage, cmap='RdYlGn_r', ax=ax, linewidths=0.5,
            linecolor='gray', annot=True, fmt='.1f', cbar=True,
            annot_kws={"size": 8})
ax.set_title('Rol vs Zona', fontsize=12)
ax.set_xlabel('Zona', fontsize=10)
ax.set_ylabel('Rol', fontsize=10)
ax.tick_params(axis='x', labelrotation=45)
ax.tick_params(axis='y', labelrotation=0)
plt.tight_layout()
plt.show()
```

Este mapa de calor muestra la distribución conjunta de los roles involucrados en siniestros viales según la zona geográfica.

* El punto más destacado (mayor densidad) se encuentra en la intersección entre conductores y zona urbana, que concentra más del 50% de las observaciones. Este es el “punto caliente” del gráfico, lo que refleja la alta frecuencia de siniestros en zonas urbanas con conductores al volante, algo esperable dado el mayor volumen de tránsito en esas áreas.

* Las zonas rurales, en cambio, tienen muy baja representación en todas las combinaciones de rol, con porcentajes menores al 1% en cada caso. Esto sugiere que los siniestros en áreas rurales son poco frecuentes en el conjunto de datos, probablemente por menor densidad vehicular.

* Las zonas suburbanas tienen participación intermedia, destacándose un poco más en los siniestros con conductores (10.4%) y pasajeros (3.7%).

Este gráfico permite ver claramente cómo los siniestros se concentran espacialmente, y cómo ciertos roles (como el del conductor) dominan la estadística en zonas de mayor densidad.

## 1. ¿Qué días y horas presentan más siniestros?

Los **miércoles** concentran la mayor cantidad de siniestros (hasta un 14,1 % en zonas urbanas), seguidos de lunes, martes y sábados.  
Los fines de semana no muestran un aumento significativo; por el contrario, los valores tienden a disminuir, lo que sugiere una fuerte relación con la movilidad laboral.

En cuanto al momento del día, la **tarde** (32,9 %) y la **noche** (23,9 %) son las franjas horarias con mayor incidencia. La **madrugada** tiene la menor cantidad de siniestros.

Además, los picos se concentran **fuera del horario pico** (definido entre las 7–9 y 17–19 h), lo cual se explica por el hecho de que las horas no pico comprenden una proporción mucho más extensa del día.

Respecto a la evolución anual, no se detecta una tendencia sostenida de aumento o disminución. Si bien hay picos puntuales, no coinciden sistemáticamente con fechas especiales como feriados.

---

## 2. ¿Cuáles son las zonas de mayor concentración?

Según los datos, la **zona urbana** es ampliamente la más afectada, concentrando la mayoría de los siniestros, en especial las colisiones entre vehículos.

Con los mapas, se identificaron varios puntos críticos:

- **Avenida Italia**, debido al alto flujo en horas pico.  
- **Villa Española**, **Paso Molino**, **Ciudad Vieja** y **Carrasco Norte**, donde se registran más siniestros con heridos graves.  
- Las **zonas suburbanas** muestran una incidencia moderada pero constante.  
- Las **zonas rurales**, en cambio, presentan una muy baja siniestralidad general.  
  Aunque existía una hipótesis de que allí los siniestros serían más graves, los datos no lo confirman claramente (apenas 1 % de casos graves en zonas rurales).

---

## 3. ¿Cómo se relacionan roles, género y edad con los siniestros?

### Según el rol:
- **Conductores**: son el grupo más involucrado, de forma consistente en todas las zonas, días y horarios.  
- **Pasajeros**: presentan menor participación y una distribución más estable.  
- **Peatones**: tienen menor frecuencia, pero una mayor proporción de siniestros graves.

### Según el género:
- Los **hombres** protagonizan la mayoría de los siniestros (52,7 % en zonas urbanas), especialmente como conductores y usuarios de birrodados.  
- Las **mujeres** aparecen con mayor frecuencia como pasajeras o involucradas en caídas y atropellos.  
- En todos los momentos del día, los hombres presentan mayor siniestralidad, con un pico los miércoles (10,7 %).

### Según la edad:
- No se cuenta con visualizaciones segmentadas por edad, pero se infiere que los **hombres adultos jóvenes** son el grupo más expuesto, debido a la alta participación en conducción de motos y vehículos livianos.

---

## 4. ¿Qué variables inciden en la gravedad del siniestro?

- La mayoría de los siniestros resultan en **heridos leves** (72,6 %), principalmente en colisiones y atropellos.  
- Los **heridos graves** se asocian sobre todo a colisiones (7,9 %) y atropellos (2,1 %).  
- Las **muertes** son poco frecuentes y están dispersas.  
- La gravedad del siniestro se relaciona con el tipo de siniestro y el tipo de vehículo, particularmente los **birrodados**.  
- No hay evidencia suficiente para afirmar que las zonas rurales concentran siniestros más graves.

---

## 5. ¿La zona influye en la cantidad o gravedad?

Sí, aunque de manera matizada.

- **Zonas urbanas**: alta frecuencia de siniestros, con predominancia de heridos leves.  
- **Zonas suburbanas**: incidencia moderada y lesiones leves.  
- **Zonas rurales**: muy baja frecuencia general y escasa representación de casos graves, lo que refuta parcialmente la hipótesis inicial sobre su mayor gravedad.

En conclusión, la **gravedad** parece estar más asociada al **tipo de siniestro y al tipo de vehículo** que a la zona en sí.

# Síntesis de hallazgos y recomendaciones para políticas públicas en Montevideo

A partir del análisis de siniestros de tránsito presentado anteriormente, es posible trazar **líneas de acción concretas y localizadas** que permitan reducir los riesgos en la ciudad. Las políticas públicas sugeridas se estructuran en torno a **cuatro ejes fundamentales**:

---

## 1. Zonas críticas: intervención urgente

Los datos muestran una alta concentración de siniestros en determinadas zonas urbanas, especialmente en:

- **Avenida Italia**, principal vía de entrada y salida de Montevideo.  
- Barrios como **Villa Española**, **Paso Molino**, **Ciudad Vieja**, **Carrasco Norte**, **Centro** y **Cordón**, que presentan altos niveles de siniestros, incluyendo casos con heridos graves.

**Acciones recomendadas:**

- Mejoras en **señalización, iluminación y visibilidad**.  
- **Rediseño vial** en cruces conflictivos o zonas de alto flujo vehicular.  
- Mayor **control de velocidades** y cumplimiento de normas en estos puntos.

---

## 2. Horarios de riesgo: medidas focalizadas en momentos clave

El análisis temporal revela que los **miércoles** son el día de mayor siniestralidad, y que las **tardes** (seguido por la noche) son las franjas más críticas, coincidiendo parcialmente con la movilidad laboral.

**Acciones recomendadas:**

- **Campañas de prevención y concientización** focalizadas en estos horarios.  
- **Refuerzo de controles y monitoreo**, especialmente entre semana por la tarde.  
- Evaluación del **diseño de semáforos y pasos peatonales** en horas críticas.

---

## 3. Grupos vulnerables: foco en los más expuestos

El perfil de los involucrados permite identificar grupos con mayor riesgo:

- **Hombres conductores de birrodados** (motocicletas, ciclomotores, triciclos, bicicletas) son quienes protagonizan la mayoría de los siniestros, y en particular los más graves.  
- **Peatones**, aunque menos numerosos, presentan una **alta proporción de lesiones graves**, especialmente en zonas céntricas.

**Acciones recomendadas:**

- **Educación vial segmentada**, con foco en **hombres jóvenes** y usuarios de vehículos de dos ruedas.  
- Reforzar la **infraestructura peatonal**, protección en cruces y señalización clara en zonas con alto tránsito de personas.  
- Promover el uso de **elementos de seguridad obligatorios** (casco, chaleco, luces) para ciclistas y motociclistas.

---

## 4. Políticas diferenciadas e integración urbana

Los patrones identificados permiten diseñar **políticas segmentadas** y con **foco territorial**, adecuadas al perfil de cada situación.

**Segmentación sugerida:**

- Por **género**: campañas específicas para población masculina, que lidera la siniestralidad.  
- Por **franja horaria**: énfasis en la tarde y noche, cuando ocurren más accidentes.  
- Por **zona**: priorizar intervenciones en áreas urbanas densas.  
- Por **tipo de vehículo**: considerar a los birrodados como categoría crítica en planificación vial.

**Visión integral recomendada:**

- Incluir los hallazgos en el diseño de **infraestructura ciclista y peatonal**.  
- Impulsar **acciones coordinadas entre tránsito, transporte, salud y planificación urbana**.

---

## Conclusión

Este estudio permite pasar del **dato a la acción**. Montevideo tiene ante sí la oportunidad de implementar una política de movilidad más **segura, inteligente y humana**, basada en evidencia.  
Si se actúa con **foco territorial, segmentación** y una **visión urbana integrada**, es posible reducir la siniestralidad vial y construir una ciudad más **habitable para todos sus habitantes**.
